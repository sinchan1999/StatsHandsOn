## Standardization

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
total_bill=list(df['total_bill'])
total_bill
```

```python

sns.histplot(total_bill)
```

### Manual Method for understanding

```python


import numpy as np
mean=np.mean(total_bill)
std=np.std(total_bill)
```

```python
mean,std
```

```python
normalized_data=[]
for i in total_bill:
    z_score=(i-mean)/std ## z_score
    normalized_data.append(z_score)

print(normalized_data)

```

```python
sns.histplot(normalized_data)
```

## Implemention using Sklearn

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.head()
```

#### Creating an instance of the StandardScaler class

```python
scaler=StandardScaler()
scaler
```

 #### Fitting the StandardScaler to the specific features in your DataFrame .

 scaler.fit() calculates the mean and standard deviation of each feature (column) in the selected DataFrame. These statistics will be used to standardize the data. Specifically:

1: The mean of each column is computed.

2: The standard deviation of each column is computed.

```python
scaler.fit(df[['total_bill','tip']])
```

#### Transforming two cols applying Standarization Technique. Then putting it in a Dataframe with column names

 After fitting the scaler, you can transform your data using scaler.transform() or scaler.fit_transform(). This transformation standardizes the data so that each feature will have a mean of 0 and a standard deviation of 1.

 Basically, appling the Z-score on the specified cols .

```python
import pandas as pd
pd.DataFrame(scaler.transform(df[['total_bill','tip']]),columns=['total_bill','tips'])
```
